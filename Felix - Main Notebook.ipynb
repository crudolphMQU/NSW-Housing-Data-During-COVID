It would actually be more interesting to the see ratio of house sales in relation to the population size since the effect of demand will be captured better. Hence, throughout this analysis section, variables of that interest will be transformed in a new dataframe that can be used for modeling in a later section.

```python
# transform feature
transDF = pivot_postcode.copy()
transDF["SalesProp"] = round(transDF["sales_no"] / transDF["population_2016"] * 100, 2)
```

```python
# show distribution for transformed variable
plt.figure(figsize=(20,7))
sns.boxplot(x="SalesProp", data=transDF)
plt.title("Distribution of House Sales", fontsize=14)
plt.xlabel("House Sales", fontsize=12)
plt.ylabel("");
```

```python
# select those with large proportions
transDF[transDF["SalesProp"] > 1.5]
```

```python
# select those with large proportions AND high numbers
transDF[(transDF["SalesProp"] > 1.5) & transDF["sales_no"] > 100]
```

```python
# compare both distributions
fig, axs = plt.subplots(2, 1, figsize=(15, 7))

sns.histplot(data=transDF, x="SalesProp", kde=True, bins=30, color="green", ax=axs[0])
sns.histplot(data=transDF, x="sales_no", kde=True, bins=30, color="blue", ax=axs[1]);
```

There are two interesting findings after transformations:
1. The distribution of proportional sales is way more normal than absolut numbers
2. There are no postcode that can be found in the subset of high proportions AND high absolut numbers. This means that the absolut high sales numbers are only due to the population size being large, hence, do not convey much relevant information. But the ratio captures the interesting postcodes where population is small but sales activity is high.

### Income

```python
plt.figure(figsize=(15, 7))
sns.histplot(x=pivot_postcode["INCP_LOW"], bins=50, color='yellow', label='Low Income')
sns.histplot(x=pivot_postcode["INCP_MID"], bins=50, color='green', label='Mid Income')
sns.histplot(x=pivot_postcode["INCP_HIGH"], bins=50, color='blue', label='High Income')
sns.histplot(x=pivot_postcode["INCP_NEG_NIL"], bins=50, color='red', label='Neg/No Income')

plt.xlabel('Number of People in Postcode')
plt.ylabel('Number of Postcodes')
plt.title('Distribution of Number of People in Postcodes per Income Bracket')
plt.legend(loc='upper right')
plt.show();
```

```python
melt_income = pd.melt(pivot_postcode1, 
                      id_vars="postcode",
                      value_vars=['INCP_NEG_NIL', 'INCP_LOW',  'INCP_MID', 'INCP_HIGH'],
                      var_name="Income",
                      value_name="CountPeoplePerPostcode")
melt_income
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=melt_income, x="CountPeoplePerPostcode", y="Income")
plt.ylabel("");
```

It can be seen that a lot of postcodes contain a high number people that belong to the low income bracket. There is even a postcode with over 35000 people in that bracket. This reduces with growing income, while the difference between high income and negative/no income is very small, meaning that there are only few more people with high income compared to no income.

It would actually be more interesting to the see income in relation to the respective population size since this will allow to better classify the proportion of high, mid, low income areas which will be more meaningful as a characteristic.

```python
# transform feature
transDF["IncNegNilProp"] = round(transDF["INCP_NEG_NIL"] / transDF["population_2016"] * 100, 2)
```

```python
# transform feature
transDF["IncLowProp"] = round(transDF["INCP_LOW"] / transDF["population_2016"] * 100, 2)
```

```python
# transform feature
transDF["IncMidProp"] = round(transDF["INCP_MID"] / transDF["population_2016"] * 100, 2)
```

```python
# transform feature
transDF["IncHighProp"] = round(transDF["INCP_HIGH"] / transDF["population_2016"] * 100, 2)
```

```python
# create column from index
transDF["postcode"] = transDF.index
```

```python
# show distribution for transformed variable

# first melt
melt_incprop = pd.melt(transDF, 
                      id_vars="postcode",
                      value_vars=['IncNegNilProp', 'IncLowProp',  'IncMidProp', 'IncHighProp'],
                      var_name="Income",
                      value_name="IncomeProportion")

# plot
plt.figure(figsize=(15, 10))
sns.boxplot(data=melt_incprop, x="IncomeProportion", y="Income")
plt.ylabel("");
```

Again, the overall trend can be validated from the initial visualisation. However, now the differences become much clearer:
- Most postcodes contain only a small fraction of people with no income, but there are some where the proportion is quite high with up to 20-30%
- There are a lot of postcodes where the proportion of low income inhabitants is 31-43%. The range in this income bracket is much larger, and there are even two postcodes that exceed 60% of low income earners.
- For mid income, the overall proportion is lower than for low income and the range is quite tight, ranging from around 10-33%. There are two postcodes with a significant low proportion (below 10%) and one postcode with a very large proportion of around 35%.
- For high income, the overall proportion is lower than for mid income. However, there are quite a lot of postcodes that contain a large proportion of high income earners with appr. 23-36%.

### Number of People in Household

```python
# use melt to put into right format
melt_household1 = pd.melt(pivot_postcode1, 
                         id_vars="postcode",
                         value_vars=['CPRF_2', 'CPRF_3', 'CPRF_4', 'CPRF_5', 'CPRF_6+', 'CPRF_na'],
                         var_name="CountPeoplePerHH",
                         value_name="CountPeoplePerPostcode")
melt_household1
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=melt_household1, x="CountPeoplePerPostcode", y="CountPeoplePerHH")
plt.ylabel("");
```

It can be seen that the number of people counted in postcodes for each number of people per households is decreasing with a growing number of people per household. Analysis of correlation with price:
- There is a weak negative relationship between the __rent level__ and the sales price level. This implies that periods with higher rent prices tend to have lower sales prices. However, as mentioned, the relationship is rather weak.
- There is a moderate positive relationship with the amount of __total bonds__.
- There is a moderate to strong negative relationship with __10 Year Bonds__. This makes sense since in economic theory investors can choose from different asset classes. Government bonds are considered as ultra safe and hence represent a baseline of what investors expect as return since they can have the bond yield "risk free". If bond yields rise, the relative return of competing asset classes decrease, demand goes down, and price level will follow.
- There is a strong negative relationship with __2 Year Bonds__. The reasoning is the same as for 10 year bonds. It is interesting, though, that the relationship for shorter duration bonds seems stronger.
- There is a moderate to strong positive relationship with __high income__. This implies that postcode area with high sales prices tend to have a population with higher income.
- There is a strong negative relationship with __interest rates__. This is another reasonable observation from an economic point of view. When interest rates decrease, it means capital becomes cheaper i.e. it is easier to get a loan since they are related to the rate level. Hence demand for housing, financed through loans, increases, and prices adjust accordingly.
- There is a moderate positive relationship with __construction costs__. This makes sense since house prices should increase as cost for building them increases.
- Lastly, there is a strong positive relationship with __sales numbers__. It is not sure at this stage how this actually relates, but the number of sales could be viewed as a proxy for demand which would drive prices.
