### Postcode only (KEN)

#### Read and clean

```python
df_pcode = pd.read_csv('Files/Cleaned/Postcode-based/Master_Sales_Rent_2017Q4_2021Q1_pcFeatures.csv')
df_pcode.drop(['skey','rkey','year','quarter'],axis=1,inplace=True)
df_pcode
```

```python
trial = df_pcode.loc[df_pcode['dwelling_type']=='Total']
trial.drop(['dwelling_type'],axis=1,inplace=True)
trial.loc[trial.isna().any(axis=1)]
```

```python
trial.dropna()['postcode'].nunique()
```

```python
trial.dropna(inplace=True)
trial
```

```python
unstacked = trial.pivot_table(index='postcode', columns = 'time_period', 
                              values = trial.drop(['postcode','time_period'],axis=1))
unstacked.columns = [' '.join(col) for col in unstacked.columns]
unstacked
```

```python
unstacked.isna().sum()
```

#### Modeling

```python
unstacked_2 = unstacked.loc[unstacked.notna().all(axis=1)]
unstacked_2
```

```python
list(unstacked.columns)
```

```python
list(unstacked_2.dtypes)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, KFold

lr = LinearRegression()
score = cross_val_score(lr, unstacked_2.drop(['median_price 2021 Q1'],axis=1), # X
                        unstacked_2['median_price 2021 Q1'],                   # y 
                        cv=KFold(n_splits=10, shuffle=True)) 
score
```
